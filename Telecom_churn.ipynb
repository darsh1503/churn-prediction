#### Encoding

```python
df.head()
```

```python
df['Payment Method'].value_counts()
```

```python
# There are two main types of encoding

# 1) one hot encoding: when we have no magnitudal relationship between options
# 2) label encoding: when we have magnitudal relationship between options
```

```python
internet_related_object_columns
```

```python
one_hot_columns = ['Offer','Multiple Lines','Payment Method'] + internet_related_object_columns
```

```python
one_hot_columns
```

```python
# Loop for encoding every column in our one hot columns list

for column in one_hot_columns:
    
    df = pd.concat([df,pd.get_dummies(df[column],prefix=column)],axis=1)
```

```python
df.head(10)
```

```python
# Since one-hot encoding will create new columns, we need to drop existing, original features

df = df.drop(columns = one_hot_columns)
```

```python
df.head()
```

```python
# label encoding - always try to do label encoding by yourselves rather than using a library
# With this method, you will have control all the encodings and manipulate them as you want
```

```python
# We will mainly use numpy's 'where' function for encodings. For example in the line below, 
# the library will look at the 'Gender' column, if it is 'Female' then replace with 1, else, 
# replace with 0

df['Gender'] = np.where(df['Gender'] == 'Female',1,0)
```

```python
df.head()
```

```python
# Do the other encodings in the same manner

df['Married'] = np.where(df['Married'] == 'Yes',1,0)
df['Phone Service'] = np.where(df['Phone Service'] == 'Yes',1,0)
df['Internet Service'] = np.where(df['Internet Service'] == 'Yes',1,0)
df['Paperless Billing'] = np.where(df['Paperless Billing'] == 'Yes',1,0)
```

```python
df.head()
```

```python
df['Internet Type'].value_counts()
```

```python
# In here, I think that the internet speed can be counted as a 'magnitudal relationship'. We know 
# that fiber is faster than Cable, Cable is faster than DSL. However, this is not a rule of thumb,
# just an interpretation, so, if you find this does not make sense, then you can apply one hot
# encoding for internet type too and try out!

df['Internet Type'] = np.where(df['Internet Type'] == 'Not Applicable',0,
                              np.where(df['Internet Type'] == 'DSL',1,
                                      np.where(df['Internet Type'] == 'Cable',2,3)))
```

```python
df.head()
```

```python
df['Contract'].value_counts()
```

```python
# Again, since the periods are shorter - longer than each other, we can assume a magnitudal relationship,
# month-to-month will be more closer to 'One Year' Contract than the 'Two Year' Contract type.

df['Contract'] = np.where(df['Contract'] == 'Month-to-Month',0,
                         np.where(df['Contract'] == 'One Year',1,2))
```

```python
df = pd.read_csv(r'telecom_customer_churn.csv')
```

```python
df.head()
```

```python
df.shape
```

```python
df_cluster = df[['Age','Avg Monthly GB Download','Monthly Charge']]
```

```python
df_cluster.head(10)
```

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
```

```python
df_cluster.isna().sum()
```

```python
df_cluster = df_cluster.fillna(0)
```

```python
df_cluster.isna().sum()
```

```python
scaler
```

```python
scaled_features = scaler.fit_transform(df_cluster)
```

```python
scaled_features
```

```python
wcss = []
```

```python
for i in range(1, 11):
    
    kmeans = KMeans(n_clusters=i)
    kmeans.fit(scaled_features)
    wcss.append(kmeans.inertia_)
```

```python
wcss
```

```python
plt.figure(figsize=(10, 6))
plt.plot(range(1, 11), wcss, marker='o', linestyle='--')
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
optimal_clusters = 3
```

```python
kmeans = KMeans(n_clusters=optimal_clusters)
```

```python
cluster_labels = kmeans.fit_predict(scaled_features)
```

```python
from collections import Counter
```

```python
Counter(cluster_labels)
```

```python
silhouette_avg = silhouette_score(scaled_features, cluster_labels)

silhouette_avg
```
